# Custom Properties

**In this example:**

Create Two Custom Types:
- **`User`:** is a descendant of `Person`
- **`Context`:** is a descendant of `Organization`

Define Custom Properties w/ Domain and Ranges:
- **`contextOf`:** Applies to `Context` (domain) and can be represented by `User` or `Context`

```python
from aloeschema.validator import AloeSchemaValidator
from aloeschema import load_schema_org, registerCustomProperty, registerCustomType

schema_org = load_schema_org()

schema_org = registerCustomType(schema_org, name="User", parent="Person", properties=[])
schema_org = registerCustomType(schema_org, name="Context", parent="Organization", properties=[])
schema_org = registerCustomProperty(schema_org, name:="contextOf",  domain=["Context"], range=["User", "Context"])

schema_org = registerCustomProperty(schema_org, name="pastAction", domain=["Context", "Person"], range=["Action"])
schema_org = registerCustomProperty(schema_org, name:="planned",    domain=["Context", "Person"], range=["Action"])
schema_org = registerCustomProperty(schema_org, name:="deferred",   domain=["Context", "Person"], range=["Action"])
schema_org = registerCustomProperty(schema_org, name:="blocked",    domain=["Context", "Person"], range=["Action"])
schema_org = registerCustomProperty(schema_org, name:="pending",    domain=["Context", "Person"], range=["Action"])
schema_org = registerCustomProperty(schema_org, name:="active",     domain=["Context", "Person"], range=["Action"])
schema_org = registerCustomProperty(schema_org, name:="cancelled",  domain=["Context", "Person"], range=["Action"])
schema_org = registerCustomProperty(schema_org, name:="completed",  domain=["Context", "Person"], range=["Action"])

schema_validator =  AloeSchemaValidator(schema_org)

print(f"""
Custom Property: `pastAction`
Description: {schema_validator._getProperty("pastAction", ignore_case=True)}

Custom Property: `planned`
Description: {schema_validator._getProperty("planned", ignore_case=True)}

Custom Property: `pending`
Description: {schema_validator._getProperty("pending", ignore_case=True)}

Custom Property: `active`
Description: {schema_validator._getProperty("active", ignore_case=True)}

Custom Property: `deferred`
Description: {schema_validator._getProperty("deferred", ignore_case=True)}

Custom Property: `blocked`
Description: {schema_validator._getProperty("blocked", ignore_case=True)}

Custom Property: `cancelled`
Description: {schema_validator._getProperty("cancelled", ignore_case=True)}

Custom Property: `completed`
Description: {schema_validator._getProperty("completed", ignore_case=True)}

Example Validations:
- person -> potentialAction -> joinaction: {schema_validator.Validate(subject_type_name="person", property_type_name="potentialAction", object_type_name="joinaction")}
- person -> pastAction -> communicateaction: {schema_validator.Validate(subject_type_name="person", property_type_name="pastAction", object_type_name="communicateaction")}
- person -> planned -> interactaction: {schema_validator.Validate(subject_type_name="person", property_type_name="planned", object_type_name="interactaction")}
- person -> pending -> transferaction: {schema_validator.Validate(subject_type_name="person", property_type_name="pending", object_type_name="transferaction")}
- person -> active -> giveaction: {schema_validator.Validate(subject_type_name="person", property_type_name="active", object_type_name="giveaction")}
- person -> deferred -> chooseaction: {schema_validator.Validate(subject_type_name="person", property_type_name="deferred", object_type_name="chooseaction")}
- person -> blocked -> rentaction: {schema_validator.Validate(subject_type_name="person", property_type_name="blocked", object_type_name="rentaction")}
- user -> cancelled -> deleteaction: {schema_validator.Validate(subject_type_name="user", property_type_name="cancelled", object_type_name="deleteaction")}
- person -> completed -> orderaction: {schema_validator.Validate(subject_type_name="person", property_type_name="completed", object_type_name="orderaction")}
""")
```
